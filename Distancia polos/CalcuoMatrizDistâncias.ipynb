```python
import pandas as pd
import geopandas
from geobr import read_municipal_seat
import geobr
```

```python
#Carregando as bases de dados
polos_saude_df = pd.read_csv('polos_AC_REGIC2018.csv')
```

```python
imedreg = geobr.read_immediate_region()
imedreg.rename(columns={'code_immediate':'CD_GEOCODI'}, inplace=True)
```

```python
## Importando de geobr as localizações municipais
#sedes_municipais = read_municipal_seat(year=2010)
#sedes_municipais.rename(columns={'code_muni':'CD_GEOCODI'}, inplace=True)
```

```python
## Criando colunas independentes para latitude e longitude
sedes_municipais = imedreg
sedes_municipais['X']=sedes_municipais.apply(lambda x:x.geometry.centroid.x, axis=1)
sedes_municipais['Y']=sedes_municipais.apply(lambda x:x.geometry.centroid.y, axis=1)
sedes_municipais
```

```python
## Filtrando colunas que vamos precisar
sedes_municipais = sedes_municipais[['name_immediate', 'CD_GEOCODI','X','Y']].copy()
sedes_municipais
```

```python
## Unindo o dataframes de municipios polo com o de coordenadas dos município
#polos_municipios_df = pd.merge(polos_municipios_df,sedes_municipais, how = 'inner')
##OBS: por conta do parâmetro how="inner", linhas com da coluna CD_GEOCODI que não aparecerem nas duas tabelas serão excluídas
##Como a nova tabela de polos municipais possui 441 linhas, nenhum polo foi excluído 

polos_municipios_df = sedes_municipais
```

```python
## Criando dataframe vazio com as linhas e colunas necessárias
df_distancias = pd.DataFrame(index = polos_municipios_df.name_immediate, 
                             columns = polos_saude_df.Municipio)

df_distancias
```

```python
##LEMBRE-SE: nas tabelas, “x” = long e “y” = lat
##Função para medir distâncias pela biblioteca GEOPY
import geopy.distance

def calc_dist(lat1, long1, lat2, long2):
  coords_1 = (lat1, long1)  
  coords_2 = (lat2, long2)  
  return (geopy.distance.distance(coords_1, coords_2).km)  ##retorna a distância em quilômetros


####EXEMPLO DE USO

##Coordenadas do ICEX
x1 = -19.8701472
y1 = -43.9628217
##Cordenadas do Mineirão
x2 = -19.8673081
y2 = -43.9722844

calc_dist(x1, y1, x2, y2)
```

```python
### PREENCHENDO O NOVO DATAFRAME
### Percorre as linhas e as colunas de "df_distancias" preenchendo os valores de acordo com as coordenadas das tabelas originais

for index, row in df_distancias.iterrows():

  index_mun_origin = polos_municipios_df.loc[polos_municipios_df["name_immediate"] == index].index[0]  ##indice do municipio na tabela original
  x_mun = polos_municipios_df.loc[index_mun_origin, "X"]
  y_mun = polos_municipios_df.loc[index_mun_origin, "Y"]

  for cidade in row.index:
 
    index_sau_origin = polos_saude_df.loc[polos_saude_df["Municipio"] == cidade].index[0] ##indice da unidade de saude na tabela original
    x_sau = polos_saude_df.loc[index_sau_origin, "X"]
    y_sau = polos_saude_df.loc[index_sau_origin, "Y"]

    df_distancias.loc[index, cidade] = calc_dist(y_mun, x_mun, y_sau, x_sau) ##LEMBRE-SE: nas tabelas, “x” = long e “y” = lat

df_distancias
#PODE DEMORAR UM POUCO
```

```python
#Salvando em CSV 
df_distancias.to_csv("matriz_distâncias.csv")
```
